```python
#F3
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import statistics
import seaborn as sns

def get_scores(path):
    scores = []
    for dirpath, dirnames, filenames in os.walk(path):
        if "pop.csv" in filenames and "/001" not in dirpath:
            file_path = os.path.join(dirpath, "pop.csv")
            df = pd.read_csv(file_path)
            scores.append(df["score_AVE"].iloc[-1])
    return scores

def ge_maxs(lst):
    return max(lst)


markov_scores = get_scores("postcanada/r100HeteroHomo/r100all_C6__CMR_0.0__IA1_20__BRN_Markov__BRN2_CGP")
CGP_scores = get_scores("postcanada/r100HeteroHomo/r100all_C8__CMR_0.0__IA1_20__BRN_CGP__BRN2_Markov")
RNN_scores = get_scores("postcanada/r100HeteroHomo/r100all_C7__CMR_0.0__IA1_20__BRN_RNN__BRN2_CGP")
markovCGP_scores = get_scores("postcanada/r100HeteroHomo/r100all_C0__CMR_0.0__IA1_10__BRN_Markov__BRN2_CGP")
RNNCGP_scores = get_scores("postcanada/r100HeteroHomo/r100all_C2__CMR_0.0__IA1_10__BRN_RNN__BRN2_CGP")
markovRNN_scores = get_scores("postcanada/r100HeteroHomo/r100all_C4__CMR_0.0__IA1_10__BRN_Markov__BRN2_RNN")
markovCGPEvolved_scores = get_scores("postcanada/r100HeteroHomo/r100all_C1__CMR_0.3__IA1_10__BRN_Markov__BRN2_CGP")
RNNCGPEvolved_scores = get_scores("postcanada/r100HeteroHomo/r100all_C3__CMR_0.3__IA1_10__BRN_RNN__BRN2_CGP")
markovRNNEvolved_scores = get_scores("postcanada/r100HeteroHomo/r100all_C5__CMR_0.3__IA1_10__BRN_Markov__BRN2_RNN")

max_values = [max(markov_scores), max(CGP_scores), max(RNN_scores), max(markovCGP_scores), max(RNNCGP_scores), max(markovRNN_scores), max(markovCGPEvolved_scores), max(RNNCGPEvolved_scores), max(markovRNNEvolved_scores)]

labels = ("Homogeneous\nMarkov", "Homogeneous\nCGP", "Homogeneous\nRNN", "50/50\nMarkov/CGP", "50/50\nRNN/CGP", "50/50\nMarkov/RNN", "Evolved\nMarkov/CGP", "Evolved\nRNN/CGP", "Evolved\nMarkov/RNN")
plt.figure(figsize=(14, 6))
plt.bar(range(len(max_values)), max_values, tick_label=labels, color='grey')
plt.xticks(ticks=range(len(max_values)), labels=labels)
#plt.xlabel('List Index')
plt.ylabel('Score')
plt.title('Maximum Scores on non-50/50 environment')
plt.show()
```
